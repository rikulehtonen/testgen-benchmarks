```python
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams.update({'font.size': 12})

# Replace 'your_file.csv' with the path to your actual CSV file
file_path = 'ppo_pretraining.csv'

# Load data from CSV file
df = pd.read_csv(file_path)

# Set the maximum step you want to display
max_step = 400

# Filter the dataframe to include only rows with 'Step' less than or equal to max_step
df_filtered = df[df["Step"] <= max_step]

# Define lighter color for the fill area
lighter_color = '#ADD8E6'  # Light blue color, you can adjust as needed

# Plotting the data
plt.figure(figsize=(10, 6))
plt.plot(df_filtered["Step"], df_filtered["Group: ppo_pretraining - ep_reward"], color="#ff8c00")
plt.fill_between(df_filtered["Step"], df_filtered["Group: ppo_pretraining - ep_reward__MIN"], df_filtered["Group: ppo_pretraining - ep_reward__MAX"], color="#fed8b1")
plt.xlabel("Episode")
plt.ylabel("Reward")
plt.title("PPO - Pre-training Reward")
plt.legend()
plt.grid(True)

# Save the plot as an EPS file
plt.savefig("ppo_pretrain.eps", format='eps', bbox_inches='tight')

# Display the plot
plt.show()
```
